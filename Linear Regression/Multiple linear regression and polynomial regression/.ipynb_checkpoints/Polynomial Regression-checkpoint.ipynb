Linear regression can still be used with polynomial problems

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data description - Col 1 = X, Col2 = Y
X = []
Y = []

for line in open('../data/data_poly.csv'):
    x, y = line.split(',')
    
    # Input is created differently for polynomial with the bias term first
    x = float(x)
    X.append([1, float(x), float(x*x)])
    Y.append(float(y))
    
# Turn into np arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Plot data
plt.scatter(X[:,1], Y)   # Since the original x is in X[1]
```

```python
# Calculate weights (similar to multiple linear regression with the difference being how X was created - Equations the same
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

```python
# Plot everything together
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yhat)

# Weird lines are due to unsorted lines
```

```python
# Correct way to plot, by sorting the points:
plt.scatter(X[:,1], Y)
plt.plot(sorted(X[:,1]), sorted(Yhat))
```

```python
# computer r-squared - Same equations as before, the difference here is that Yhat was calculated differently above from 1D
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("r-squared:", r2)
```
