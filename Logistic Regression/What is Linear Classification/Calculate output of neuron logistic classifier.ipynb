```python
import numpy as np

# demonstrates how to calculate the output of a logistic unit using numpy.
# the data X and weight matrix w are randomly generated from a
# standard normal distribution.
N = 100
D = 2
```

```python
X = np.random.randn(N,D)
# Add bias terms
# Array in numpy is 1d, we need 2d, so we do it like this
ones = np.ones((N, 1))
# Concatinated to original X
Xb = np.concatenate((ones, X), axis=1)
```

```python
Xb
```

```python
# Weight vector
w = np.random.randn(D + 1)
```

```python
# Get dot product between each X and w
z = Xb.dot(w)
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
print(sigmoid(z))
```
